```python
data = data[["Car_Owner",'Propert_Owner','Annual_income','EDUCATION','label']]
data
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
data["Car_Owner"] = labelencoder.fit_transform(data["Car_Owner"])
data["Propert_Owner"] = labelencoder.fit_transform(data["Propert_Owner"])
data["EDUCATION"] = labelencoder.fit_transform(data["EDUCATION"])
```

```python
data
```

```python
data = data[["Car_Owner",'Propert_Owner','Annual_income','EDUCATION','label']]
data
```

```python
x = data.iloc[: , : -1]
x
```

```python
y = data.iloc[:, -1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler_x_train = scaler.fit_transform(x_train)
scaler_x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelperformance(prediction):
    print("accuracy score on medel is {}". format(accuracy_score(y_test, prediction)))
```

```python
from sklearn.linear_model import LogisticRegression
regression = LogisticRegression()
regression.fit(scaler_x_train,y_train)
```

```python
log_pred = regression.predict(scaler_x_test)
```

```python
modelperformance(log_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_neighbors": [1, 3, 5, 7, 9, 11, 13],
    "weights": ["uniform", "distance"],
    "algorithm": ["auto", "kd_tree"]
}

gridkn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=2)
```

```python
gridkn.fit(scaler_x_train,y_train)
```

```python
gridkn.best_params_
```

```python
knpreds = gridkn.predict(scaler_x_test)
```

```python
modelperformance(knpreds)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svm = SVC()
param_grid_svr = {
    "C": [0.01, 0.1, 0.5, 1],
    "kernel": ["linear", "rbf", "poly"]
}

gridsvr = GridSearchCV(svm, param_grid_svr, cv=5)
```

```python
gridsvr.fit(scaler_x_train,y_train)
```

```python
svmpreds = gridsvr.predict(scaler_x_test)
```

```python
modelperformance(svmpreds)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    "criterion": ["gini", "entropy"],
    "splitter": ["best", "random"],
    "max_depth": [None, 10, 20, 30, 40, 50],
    "min_samples_split": [2, 5, 10, 15],
    "min_samples_leaf": [1, 2, 4]
}

dtc = DecisionTreeClassifier()
grid_dtc = GridSearchCV(dtc, param_grid, cv=5)
```

```python
grid_tree = GridSearchCV(estimator = DecisionTreeClassifier(), param_grid = param_grid)
```

```python
grid_tree.fit(scaler_x_train,y_train)
```

```python
treepred = grid_tree.predict(scaler_x_test)
```

```python
modelperformance(treepred)
```
